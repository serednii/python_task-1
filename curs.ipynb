```python
import re
print("Please set memory size and max output width:")
# sizeMemory, maxOutput = input("Please set memory size and max output width:").split()
# sizeMemory = int(sizeMemory)
# maxOutput = int(maxOutput)
exitFlag = True
sizeMemory = 30
maxOutput = 10
memoryIndexAndSize = [[0, sizeMemory, 'free']]
# memoryIndexAndSize = [[0,5, 'busy'],[5,3, 'free'],[8,4, 'free'],[12,5, 'busy'], [17, 5, 'free'],  [22, 4, 'free'],  [26, 4, 'free']]
print(memoryIndexAndSize)


def findIdMemoryBlock(list, elementId):
    for row in list:
        if row[0] == elementId:
            return True  # Знайдено шуканий елемент
    return False  # Шуканий елемент не знайдено 

def sprintHelp():
        print()
        print('************************************HELP*******************************************************')
        print('                  help  - show this help') 
        print('                  exit  - exit this program')
        print('                  print - print memory blocks map')
        print('                  allocate <num> - allocate <num> cells. Returns block first cell number')
        print('                  free <num> - free block with first cell number <num>')
        print('***********************************************************************************************')
        print()

def findFreeMemoryBlock(list, size):
    print()
    newList = []
    for element in list:
        if element[2] == 'free' & element[1] >= size:
            if element[1] == size:
                element[2] = 'busy'
            else:
                _index, _size, _status = list  

                print(_index)
                print(_size)
                print(_status)
                newList.append([_index, size, 'busy'])
                newList.append(_index + size, _size-size, 'free')
        else:
            newList.append(element)

def deleteMemoryBlock(list, elementId):
    count = 0
    sizeList = len(list)
    numberFreeElement = 0 #індекс підмасива в якому звільнено память
    while count < sizeList:
        if list[count][0] == elementId:
            list[count][2] = 'free'
            numberFreeElement = count
        count+=1
    return [list, numberFreeElement]  

def joinFreePrevElements(list, index): #індекс підмасива в якому звільнили память
    _statusPrev = list[index-1][2] if index > 0 else 'busy'
    if _statusPrev == 'busy':
        return [list, index]
    elif _statusPrev == "free":
        _size = list[index][1]
        list[index-1][1] += _size # Сумуємо розміридвох вільних блоків
        list.pop(index)
        # return joinFreePrevElements(list, index-1)
    return [list,index-1]

def joinFreeNextElements(list, index):#індекс підмасива в якому звільнили память
    sizeList = len(list) 
    _indexNext, _sizeNext, _statusNext = list[index + 1] if index < sizeList-1 else [0, 0, 'busy']
    if _statusNext == 'busy':
        return [list, index]
    elif _statusNext == "free":
        list[index][1] += _sizeNext
        list.pop(index+1)
        # return joinFreeNextElements(list, index)
    return [list,index]


def removeNewBlockMemory(list, rezInput):
    blockId = int(rezInput.split()[1]) #Забираємо число
    newArray = []
    if findIdMemoryBlock(list, blockId):
        newArray, index = deleteMemoryBlock(list, blockId)
        newArray, index = joinFreePrevElements(newArray, index) #Обєднати попередній вільний блок
        newArray, index = joinFreeNextElements(newArray, index) #Обєднати наступний вільний блок
    else:
        print("Такого id не існує")
    return newArray

def addNewBlockMemory(list, rezInput):
        size = int(rezInput.split()[1]) #Забираємо число
        if size > 0:
            newList = []
            flagFindFree = False #Позначає що найшли вільний блок памяті
            for element in list:
                if element[2] == 'free'and element[1] >= size and not flagFindFree:
                    flagFindFree = True
                    if element[1] == size:
                        element[2] = 'busy'
                        newList.append(element)
                    else:
                        _index, _size, _status = element #8 4 free (size =2) деструктуризуємо список
                        newList.append([_index, size, 'busy']) #8 2 busy
                        newList.append([_index + size, _size-size, 'free']) #10 2 free добавляємо ще запис щоє вільні байти
                else:
                    newList.append(element) #блок непідходить він є або занятий або має замалий розмір вільних байт

            if not flagFindFree: #Якщо прапор не помінявся Вільної памяті не знайдено
                print('Вільної памяті не знайдено')
            return newList
        else: 
            print('Кількість комірок памяті для виділення має бути більше як ноль') 

def printDampMemory():
    tempString = ""
    #Формуємо повну строку
    for element in memoryIndexAndSize:
        if element[2] == 'busy':
            tempString += " " + str(element[0])  + (''.join(["XX"] * element[1]))
        else:
            tempString += (''.join(["__"] * element[1]))

    countStrok = int(sizeMemory / maxOutput)
    cuontSymbolStrok = int(len(tempString) / countStrok)# Визначаємо довжину строки
    startSymbol = 1
    endSymbol = cuontSymbolStrok
    while countStrok != 0:
            sss = tempString[startSymbol-1: endSymbol].strip() #Обрізаємо першу строку 
            sss = re.sub(r'\s', ' | ', sss)
            print("| "+ sss +" |")
            startSymbol+=cuontSymbolStrok
            endSymbol+=cuontSymbolStrok
            countStrok-=1


while exitFlag:
    print("Введіть команду для довідки введіть help")
    rezInput = input("Введіть команду для довідки введіть help").strip().lower()
    if rezInput == 'e':
        exitFlag = False 
    elif rezInput == 'h':
        printHelp() 
    elif rezInput == 'p':
        printDampMemory()
    elif  re.match('^a\s+\d+', rezInput):
        memoryIndexAndSize = addNewBlockMemory(memoryIndexAndSize, rezInput)
        print(memoryIndexAndSize)
    elif re.match('^f\s+\d+', rezInput): 
        memoryIndexAndSize = removeNewBlockMemory(memoryIndexAndSize, rezInput)
        print (memoryIndexAndSize)
    else:
        print("Такої команди не існує")


print('EXIT PROGRAM')


```
